## 1 Domain-specific area and objectives of the project

The chosen domain for this report is fossil fuels, specifically focusing on one of the most pivotal and widely traded commodities: **oil**.

Oil, or petroleum, has been a vital resource that has played a crucial role in the global economy for decades. Its significance as a primary source of energy, raw material for various industries, and geopolitical leverage underscores its significance worldwide. Oil is actively traded on numerous exchanges, with futures contracts serving as a primary module for investors and traders to engage in speculation regarding future price movements.

The large production of oil worldwide has also brought global attention regarding its effects on climate change. The production and consumption of fossil fuels like oil, releases tons of carbon dioxide into the atmosphere, contributing to global warming[1].

In addition to being a fundamental component of global commerce, oil is often viewed as a barometer for economic health. Fluctuations in oil prices can reflect shifts in demand, geopolitical tensions, and broader economic trends. As such, having a sense of yearly oil production and predicting its future production become paramount, especially amid the complexities of the current global conditions, supply chain disruptions, and environmental concerns.

**As of now, with global market instability, war, inflation and climate change, determining what the yearly oil production of a state or country can help governments, investors and scientists in planning of government budgets, investments, research, and can help in deciding in oil production should increase or decrease.**
